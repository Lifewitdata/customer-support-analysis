```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load synthetic workforce data (Assuming CSV file)
df = pd.read_csv("customer_support_data.csv")

# Display first few rows
print(df.head())
```

```python
# Inspect the data
print(df.info())
print(df.describe())
```

```python
# Convert timestamps to datetime
df['created_at'] = pd.to_datetime(df['created_at'])
df['resolved_at'] = pd.to_datetime(df['resolved_at'])

# Check for duplicates
print(f"Duplicates: {df.duplicated().sum()}")  # Remove if any: df = df.drop_duplicates()

# Check for missing values
print(df.isnull().sum())  # customer_rating has missing values

# Filter resolved tickets for resolution time analysis
resolved_df = df[df['status'] == 'Resolved']
```

```python
df['date'] = df['created_at'].dt.date
ticket_volume = df.groupby('date')['ticket_id'].count()

plt.figure(figsize=(12, 6))
sns.lineplot(x=ticket_volume.index, y=ticket_volume.values)
plt.title('Ticket Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Tickets')
plt.xticks(rotation=45)
plt.show()
```

```python
df['hour'] = df['created_at'].dt.hour
peak_hours = df.groupby('hour')['ticket_id'].count()

plt.figure(figsize=(10, 5))
sns.barplot(x=peak_hours.index, y=peak_hours.values)
plt.title('Peak Hours for Support Requests')
plt.xlabel('Hour of Day (0-23)')
plt.ylabel('Number of Tickets')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(resolved_df['resolution_time'], bins=30, kde=True)
plt.title('Resolution Time Distribution (Resolved Tickets)')
plt.xlabel('Resolution Time (hours)')
plt.ylabel('Frequency')
plt.show()
```

```python
agent_productivity = resolved_df.groupby('agent_id')['ticket_id'].count()

plt.figure(figsize=(12, 6))
sns.barplot(x=agent_productivity.index, y=agent_productivity.values)
plt.title('Tickets Resolved per Agent')
plt.xlabel('Agent ID')
plt.ylabel('Number of Tickets')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='resolution_time', y='customer_rating', data=resolved_df)
plt.title('Resolution Time vs. Customer Rating')
plt.xlabel('Resolution Time (hours)')
plt.ylabel('Customer Rating (1-5)')
plt.show()
```

```python
df.groupby('cluster').agg({
    'lob': lambda x: x.mode()[0],  # Get the most frequent value
    'resolution_time': 'mean'  # Get the average resolution time
})
```

```python
from statsmodels.tsa.arima.model import ARIMA

ticket_volume_ts = df.set_index('created_at').resample('D')['ticket_id'].count()
model = ARIMA(ticket_volume_ts, order=(1, 1, 1))
model_fit = model.fit()
forecast = model_fit.forecast(steps=30)

plt.figure(figsize=(12, 6))
plt.plot(ticket_volume_ts, label='Historical')
plt.plot(forecast, label='Forecast', color='red')
plt.title('Ticket Volume Forecast (Next 30 Days)')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans

le_lob = LabelEncoder()
df['lob_encoded'] = le_lob.fit_transform(df['lob'])
features = df[['lob_encoded', 'resolution_time']].fillna(0)  # Fill NaN for clustering

kmeans = KMeans(n_clusters=3)
df['cluster'] = kmeans.fit_predict(features)

print(df.groupby('cluster').agg({'lob': lambda x: x.mode().iloc[0], 'resolution_time': 'mean'}))
```
